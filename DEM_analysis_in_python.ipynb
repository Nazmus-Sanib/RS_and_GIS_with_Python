```python
"""This project is about data analysis of the Digital Elevation Model 
   of the part of the Chattagram City through the Python Scripting.
   The project is included:
   (i) Insert a scale
   (ii) Create a hillshade layer"""
   
   

import rasterio as rio
import earthpy as et
import earthpy.plot as ep
import earthpy.spatial as es
import matplotlib.pyplot as plt
import matplotlib.image as imp
```

```python
# Read and print raster-data

"""References:

https://rasterio.readthedocs.io/en/latest/quickstart.html#opening-a-dataset-in-reading-mode

https://earthpy.readthedocs.io/en/latest/gallery_vignettes/
plot_bands_functionality.html#sphx-glr-gallery-vignettes-plot-bands-functionality-py"""



with rio.open("AP_16361_FBS_F0440_RT1/AP_16361_FBS_F0440_RT1.dem.tif") as data0:
    dem = data0.read(1)
    
ep.plot_bands(dem,cbar = False)
```

```python
# Start of (i) Insert a scale

"""Reference:
https://matplotlib.org/stable/tutorials/introductory/images.html"""


im0 = imp.imread("dem_0_Ch.png")
color0 = im0[:,:,0]
fig,ax = plt.subplots(figsize = (12,20))
fig_ = ax.imshow(color0,cmap = 'nipy_spectral')
cbar_ = fig.colorbar(fig_, ticks = [i/10 for i in range(10)],orientation = 'horizontal')


# End of (i) Insert a scale
```

```python
# Start of (ii) Create a hillshade layer

"""Reference:
https://earthpy.readthedocs.io/en/latest/gallery_vignettes/
plot_dem_hillshade.html#sphx-glr-gallery-vignettes-plot-dem-hillshade-py"""




hillshade = es.hillshade(dem)

ep.plot_bands(hillshade,cbar = False, cmap = "terrain", "Hillshade_layer")
```

```python
im1 = imp.imread("hillshade.png")
fig,ax = plt.subplots(figsize = (12,20))
fig_ = ax.imshow(im1,cmap = 'nipy_spectral')

l = [i/10 for i in range(10)]

cbar_ = fig.colorbar(fig_,ticks = l,orientation = 'horizontal')
cbar_.ax.set_xticklabels(['800.0','700.0','600.0','500.0','400.0','300.0','200.0','100.0','0.0','-100.0'])


# End of (ii) Create a hillshade layer
```
